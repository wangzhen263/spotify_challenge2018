```python
import pandas as pd
import re
import pickle


def normalize_name(name):
    name = name.lower()
    name = re.sub(r"[.,\/#!$%\^\*;:{}=\_`~()@]", ' ', name)
    name = re.sub(r'\s+', ' ', name).strip()
    return name

vertices = pd.read_json('mpd.epgm/vertices.json', lines=True)
for index, row in vertices.iterrows():
    row['data']['name'] = normalize_name(row['data']['name'])
vertices.head()
```

```python
playlists = vertices.loc[vertices['meta'].apply(lambda r:r['label']=='playlist')]
playlists.shape
playlists.head()
```

```python
tracks = vertices.loc[vertices['meta'].apply(lambda r:r['label']=='track')]
tracks.shape
tracks.head()
```

```python
artists = vertices.loc[vertices['meta'].apply(lambda r:r['label']=='artist')]
artists.shape
artists.head()
```

```python
albums = vertices.loc[vertices['meta'].apply(lambda r:r['label']=='album')]
albums.shape
albums.head()
```

```python
edges = pd.read_json('mpd.epgm/edges.json', lines=True)
edges.head()
```

```python
all_playlist_to_track = edges.loc[edges['meta'].apply(lambda r:r['label']=='playlist_to_track')]
all_playlist_to_track.head()

# all_playlist_to_track.loc[all_playlist_to_track['source'].apply(lambda r:r==0)]
```

```python
all_track_to_playlist = edges.loc[edges['meta'].apply(lambda r:r['label']=='track_to_playlist')]
all_track_to_playlist.head()
```

```python
all_track_to_artist = edges.loc[edges['meta'].apply(lambda r:r['label']=='track_to_artist')]
all_track_to_artist.head()
```

```python
all_track_to_album = edges.loc[edges['meta'].apply(lambda r:r['label']=='track_to_album')]
all_track_to_album.head()
```

```python
linktracks = all_playlist_to_track.loc[all_playlist_to_track['source'].apply(lambda r:r==0)]
selected_tracks = tracks.loc[tracks['id'].isin(linktracks['target'])]
track = selected_tracks.head(1)
artist_id = all_track_to_artist.loc[all_track_to_artist['source']==track.id.values[0]]
album_id = all_track_to_album.loc[all_track_to_album['source']==track.id.values[0]]
artist_name = artists.loc[artists['id']==('spotify:artist:'+artist_id.target.values[0])]
albums_name = albums.loc[albums['id']==('spotify:album:'+album_id.target.values[0])]
albums_name['data'].values[0]['name']
# albums_name['data']['name']
```

```python
texts = []
for index, row in playlists.iterrows():
    texts.append(row['data']['name'])
    linktracks = all_playlist_to_track.loc[all_playlist_to_track['source'].apply(lambda r:r==row['id'])]
    selected_tracks = tracks.loc[tracks['id'].isin(linktracks['target'])]
    for index, track in selected_tracks.iterrows():
        texts.append(track['data']['name'])
        artist_id = all_track_to_artist.loc[all_track_to_artist['source']==track.id]
        album_id = all_track_to_album.loc[all_track_to_album['source']==track.id]
        artist_name = artists.loc[artists['id']==('spotify:artist:'+artist_id.target.values[0])]
        albums_name = albums.loc[albums['id']==('spotify:album:'+album_id.target.values[0])]
        texts.append(artist_name['data'].values[0]['name'])
        texts.append(albums_name['data'].values[0]['name'])

open('listfile.txt', 'a').write(' '.join(texts) + '\n')
```
